```python
import librosa
import IPython.display as ipd
import numpy as np
```

```python
# Load one song
X, sr = librosa.load('../data/fma_small/000/000002.mp3')
```

```python
X.shape
```

```python
# Play audio
ipd.Audio(X, rate=sr)
```

```python
# Short time fourier transforrmation
stft = np.abs(librosa.stft(X, n_fft=2048, hop_length=512))
stft.shape
```

```python
librosa.display.specshow(stft)
```
